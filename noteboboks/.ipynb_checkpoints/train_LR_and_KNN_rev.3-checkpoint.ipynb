```python
"""
Regime-change classifier (inflation) – no target leakage + L1 sparsity
======================================================================

Key changes vs. v1
------------------
1.  'Zscore' removed from features (it defined the label!).
2.  LogisticRegression now uses **L1** penalty (sparse, interpretable).
3.  `GridSearchCV` with `TimeSeriesSplit` tunes C (regularisation strength).
"""

# -------------------------------------------------------------------
# 0. Imports
# -------------------------------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer  # noqa: F401
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.metrics import (classification_report, confusion_matrix,
                             roc_auc_score, roc_curve,
                             precision_recall_curve)

# -------------------------------------------------------------------
# 1. Load dataset
# -------------------------------------------------------------------
df = (
    pd.read_csv(
        r"C:\Users\ghkjs\GitHub\regime-shift-classification\data\processed\inflation_regime_dataset.csv",
        index_col=0,
        parse_dates=True
    )
      .sort_index()
)

# -------------------------------------------------------------------
# 2.  Remove leakage column(s) & define X / y
# -------------------------------------------------------------------
X = df.drop(columns=["Inflation_Regime", "Zscore"])   # << NO Zscore >>
y = df["Inflation_Regime"]

# -------------------------------------------------------------------
# 3. Chronological split (70 / 30)
# -------------------------------------------------------------------
split_idx = int(len(df) * 0.70)
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

print(f"Train span : {X_train.index.min().date()} → {X_train.index.max().date()}  ({len(X_train)})")
print(f"Test  span : {X_test.index.min().date()}  → {X_test.index.max().date()}   ({len(X_test)})\n")

# -------------------------------------------------------------------
# 4. Shared preprocessing
# -------------------------------------------------------------------
imputer = IterativeImputer(random_state=42, max_iter=10, initial_strategy="median")
scaler  = StandardScaler()

# -------------------------------------------------------------------
# 5. Logistic Regression with L1 – hyper-parameter search
# -------------------------------------------------------------------
base_lr = LogisticRegression(
    penalty        ="l1",
    solver         ="liblinear",          # supports L1
    class_weight   ="balanced",
    max_iter       =3000,
    random_state   =42
)

pipe_lr = Pipeline([
    ("imputer", imputer),
    ("scaler",  scaler),
    ("clf",     base_lr)
])

# grid of C (inverse of regularisation strength)
param_grid = {"clf__C": [0.1, 0.3, 0.5, 1.0, 3.0, 10.0]}

tscv = TimeSeriesSplit(n_splits=5)

grid_lr = GridSearchCV(
    pipe_lr,
    param_grid,
    cv           =tscv,
    scoring      ="roc_auc",
    n_jobs       =-1,
    verbose      =1
).fit(X_train, y_train)

print(f"\nBest C from CV = {grid_lr.best_params_['clf__C']} "
      f"(mean AUC = {grid_lr.best_score_:.3f})")

# -------------------------------------------------------------------
# 6. Evaluate on hold-out test set
# -------------------------------------------------------------------
best_lr = grid_lr.best_estimator_

y_prob_lr = best_lr.predict_proba(X_test)[:, 1]
y_pred_lr = (y_prob_lr >= 0.5).astype(int)          # default 0.5 for now

print("\n=== Logistic Regression (L1, leakage-free) ===")
print(classification_report(y_test, y_pred_lr, digits=3))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_lr))
print("ROC AUC:", roc_auc_score(y_test, y_prob_lr).round(3))

# ---------------------------------------------------- ROC curve
fpr, tpr, _ = roc_curve(y_test, y_prob_lr)
plt.figure()
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, y_prob_lr):.2f}")
plt.plot([0,1],[0,1],'--')
plt.title("ROC – Logistic Regression (leakage-free)")
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.grid(True); plt.legend(); plt.show()

# -------------------------------------------------------------------
# 7. Precision–Recall optimisation (as before, but on new probs)
# -------------------------------------------------------------------
prec, rec, thr = precision_recall_curve(y_test, y_prob_lr)
f1 = 2*(prec[:-1]*rec[:-1])/(prec[:-1]+rec[:-1])
best_i  = np.argmax(f1)
best_th = thr[best_i]

print(f"\nBest F1 = {f1[best_i]:.3f} at threshold = {best_th:.3f}")

y_pred_opt = (y_prob_lr >= best_th).astype(int)
print("\n=== LR @ F1-optimal threshold ===")
print(classification_report(y_test, y_pred_opt, digits=3))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_opt))

# -------------------------------------------------------------------
# 8.  Coefficient plot (after leakage fix)
# -------------------------------------------------------------------
coef_vals   = best_lr.named_steps["clf"].coef_.ravel()
features    = X.columns
coef_df = (
    pd.DataFrame({"feature": features,
                  "coef"   : coef_vals,
                  "abs"    : np.abs(coef_vals)})
      .query("abs > 1e-6")                     # drop exact zeros from L1
      .sort_values("abs", ascending=True)
)

plt.figure(figsize=(8, 0.3*len(coef_df)+1))
plt.barh(coef_df["feature"], coef_df["coef"])
plt.axvline(0,color='k',ls='--')
plt.title("Logistic-Regression coefficients (L1, no Zscore)")
plt.xlabel("Coefficient"); plt.tight_layout(); plt.show()

# -------------------------------------------------------------------
# 9.  (Optional) KNN re-run – unchanged, but now with same X
# -------------------------------------------------------------------
knn_pipe = Pipeline([
    ("imputer", imputer),
    ("scaler",  scaler),
    ("clf",     KNeighborsClassifier(n_neighbors=5))
]).fit(X_train, y_train)

y_prob_knn = knn_pipe.predict_proba(X_test)[:,1]
y_pred_knn = (y_prob_knn >= 0.5).astype(int)

print("\n=== KNN (no leakage) ===")
print(classification_report(y_test, y_pred_knn, digits=3))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
print("ROC AUC:", roc_auc_score(y_test, y_prob_knn).round(3))
```
