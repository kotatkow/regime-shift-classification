```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# ─────────────────────────────────────────────
# 1. Load the full dataset
# ─────────────────────────────────────────────
df = pd.read_csv(
    r'C:\Users\ghkjs\GitHub\regime-shift-classification\data\processed\inflation_regime_dataset.csv',
    index_col=0,
    parse_dates=True
)

# ─────────────────────────────────────────────
# 2. Optional: Hold out last 12 months (for true test later)
# ─────────────────────────────────────────────
cutoff = df.index.max() - pd.DateOffset(months=12)
df_train = df.loc[:cutoff]

# ─────────────────────────────────────────────
# 3. Prepare training features & target, drop Zscore
# ─────────────────────────────────────────────
X_train = df_train.drop(columns=['Inflation_Regime', 'Zscore'])
y_train = df_train['Inflation_Regime']

# ─────────────────────────────────────────────
# 4. Build the logistic-regression pipeline (best hyperparams)
# ─────────────────────────────────────────────
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),      # fill missing values
    ('scaler',  StandardScaler()),                      # standardize features
    ('clf',     LogisticRegression(
        C=100,
        penalty='l2',
        solver='saga',
        class_weight=None,
        max_iter=2000
    ))
])

# ─────────────────────────────────────────────
# 5. Fit the model
# ─────────────────────────────────────────────
pipeline.fit(X_train, y_train)

# ─────────────────────────────────────────────
# 6. Extract & display coefficients
# ─────────────────────────────────────────────
feature_names = X_train.columns
coefs = pipeline.named_steps['clf'].coef_[0]

coef_df = (
    pd.DataFrame({'feature': feature_names, 'coef': coefs})
      .sort_values('coef', ascending=False)
)

print("Logistic Regression Coefficients (without Zscore):")
print(coef_df.to_string(index=False))

# ─────────────────────────────────────────────
# 7. Plot the coefficients
# ─────────────────────────────────────────────
plt.figure(figsize=(8, 5))
coef_df.set_index('feature')['coef'].plot(kind='barh')
plt.title('Logistic Regression Coefficients\n(Positive → High-Inflation)')
plt.xlabel('Coefficient Value')
plt.grid(True)
plt.tight_layout()

# ─────────────────────────────────────────────
# 8. Save the figure
# ─────────────────────────────────────────────
output_path = r'C:\Users\ghkjs\GitHub\regime-shift-classification\results\coefficients.png'
plt.savefig(output_path, dpi=300)
print(f"Saved coefficient plot to: {output_path}")

# Finally, display it
plt.show()
```
